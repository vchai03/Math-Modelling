```python
e = 2.71828

A = np.matrix([[1,0]]) # percentages
P = np.matrix([[1,0],[1,0]]) # transition matrix

bias = -1150000

# cost in $/mile in 2020
gas_cost = 0.49316146
electric_cost = 0.2216

# change in $ for cost per year
delta_gas_cost = .0297
delta_electric_cost = -0.01085338

tot_miles = 1110000

electric_truck = 180000 # cost of electric truck
diesel_truck = 115000 # cost of normal truck

time = 878.666 # difference in hours of wasted time over 12
salary = 21 # hourly wage in dollars

years = []
percentage_electric = []
k = .000008
A = np.matrix([[1,0]]) # percentages

for i in range(21):

  C1 = (gas_cost + delta_gas_cost * i - electric_cost - delta_electric_cost * i) * tot_miles # calculate operating cost difference
  C2 = diesel_truck - electric_truck # calculate purchasing cost difference
  C3 = time * salary # calculate opportunity cost

  Total = C1 + C2 + C3 + bias
  x = -k * Total

  #calculate sigmoid function
  prob = 1/(1 + pow(e,x))

  #transition matrix
  P = np.matrix([[1-prob,prob],[0,1]])

  A = np.matmul(A,P)

  #store values for graphing
  years.append(2020+i)
  percentage_electric.append(A[0,1]*100)

  plt.plot(years,percentage_electric)
  plt.xlabel('Year')
  plt.ylabel('Percentage of Electric Trucks')
  plt.title('Percentage of Electric Trucks Over Time')
  plt.show()
```
